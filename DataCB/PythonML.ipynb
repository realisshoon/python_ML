#Numpy
```Numerical Python```을 의미하며 파이썬에서 선형대수 기반의 프로그램을 쉽게 만들 수 있도록 지원하는 패키지이다. 배열 기반의 연산과 다양한 데이터 핸들링 기능을 제공한다.

##Numpy ndarray

````python
import numpy as np 
````

numpy 기반 데이터 타입은 ndarray이다. 다차원 배열을 쉽게 생성하고 다양한 연산을 수행 할 수 있다.

````python
array1= np.array([1,2,3])
print('array1 type:',type(array1))
print('array1 array 형태: ', array1.shape)

array2=np.array([[1,2,3],[2,3,4]])
print('array2 type:',type(array2))
print('array2 array 형태: ', array2.shape)


array3=np.array([[1,2,3]])
print('array3 type:',type(array3))
print('array3 array 형태: ', array3.shape)
````

```shape```을 통해 각 배열의 row와 column을 파악 할 수 있다. 데이터 세트 간의 입출력과 변환을 수행하다보면 데이터 간의 차원이 달라서 오류가 발생 할 수 있는데 이때는 ```reshape()```를 이용하여 차원을 변환 시켜 줄 수 있다.
###ndaarya.ndim
차원을 확인 할 수 있다.

````python
print('array1: {:0}차원, array2: {:1}차원, array3: {:3}차원'.format(array1.ndim,array2.ndim,array3.ndim))
````

##붓꽃 품종 예측하기
- 지도학습의 대표적인 방법 중 하나인 분류에 대해서 공부하기 위해 붓꽃 데이터 세트를 `sklearn.datasets`내 에서 불러옵니다.

````python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier #Decision Tree 알고리즘
from sklearn.model_selection import train_test_split #학습 데이터와 테스트 데이터를 분리하는 함수
````

````python
import pandas as pd

iris=load_iris() #붓꽃 데이터 세트를 로딩

iris_data=iris.data 

iris_label=iris.target
print('iris target 값: ', iris_label)
print('iris target 명: ', iris.target_names)

#붓꽃 데이터 세트를 자세히 보기 위해 DataFrame으로 변환
iris_df=pd.DataFrame(data=iris_data,columns=iris.feature_names)
iris_df['label']=iris.target
iris_df.head(3)
````
